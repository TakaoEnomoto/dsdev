```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./sample_data/california_housing_train.csv")
```

```python
pop_data = np.array(data["population"])
```

```python
print(f"合計値：{np.sum(pop_data)}")
print(f"平均値：{np.mean(pop_data)}")
print(f"中央値：{np.median(pop_data)}")
print(f"標本分散：{np.var(pop_data, ddof=0)}")
print(f"不偏分散：{np.var(pop_data, ddof=1)}")
print(f"標本標準偏差：{np.std(pop_data, ddof=0)}")
print(f"不偏標準偏差：{np.std(pop_data, ddof=1)}")
```
