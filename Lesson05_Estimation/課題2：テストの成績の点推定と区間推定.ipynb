```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_theme()
```

```python
data = pd.read_csv("students_scores.csv")
```

```python
eng_score = data["english"]
```

```python
plt.hist(eng_score)
```

```python
sns.histplot(eng_score, kde=True)
```

```python
mu = np.mean(eng_score)
mu
```

```python
sigma2 = np.var(eng_score, ddof=1)
sigma2
```

```python
sigma = np.std(eng_score, ddof=1)
se = sigma / np.sqrt(len(eng_score))
se
```

```python
df = len(eng_score) - 1

interval = stats.t.interval(0.95, df, loc=mu, scale=se)
interval
```
